```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
#Load SQLite database
conn = sqlite3.connect("/Users/Maniosh/Desktop/sales_data.db")

#Run basic SQL: query

query = """
SELECT 
    product, 
    SUM(quantity) AS total_quantity, 
    SUM(quantity * price) AS revenue
FROM sales
GROUP BY product;
"""
#load into pandas

df = pd.read_sql_query(query, conn)

#Print results:

print("Sales Summary:")
print(df)

#plot bar chart

df.plot(kind='bar', x='product', y='revenue')

plt.title("Revenue by Product")
plt.xlabel("Product")
plt.ylabel("Revenue")

plt.tight_layout()
plt.show()

plt.savefig("sales_chart.png")
conn.close()
```

```python
# Find highest revenue product using pandas
df_sorted = df.sort_values(by="revenue", ascending=False)

print("Highest Revenue Product:")
print(df_sorted.head(1))
```
